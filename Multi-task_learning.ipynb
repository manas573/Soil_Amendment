```python
from tensorflow.keras.layers import InputLayer
from keras.layers import Input
from keras.models import Model
from tensorflow.keras.utils import plot_model
n_features = x.shape[1]

y_class = LabelEncoder().fit_transform(y[:,0])
n_class = len(np.unique(y_class))

# split data into train and test sets
X_train, X_test, y_train, y_test, y_train_class, y_test_class = train_test_split(x, y[:,1], y_class, test_size=0.40, random_state=1)
# input

def build_model(hp):
    visible = Input(shape=(n_features,))
    hidden1 = Dense(hp.Int("units", min_value=1, max_value=500, step=1), activation=hp.Choice("activation", ["relu", "tanh","selu"]), kernel_initializer='he_normal')(visible)
    hidden2 = Dense(hp.Int("units", min_value=1, max_value=500, step=1), activation=hp.Choice("activation", ["relu", "tanh","selu"]), kernel_initializer='he_normal')(hidden1)
    hidden3 = Dense(hp.Int("units", min_value=1, max_value=500, step=1), activation=hp.Choice("activation", ["relu", "tanh","selu"]), kernel_initializer='he_normal')(hidden2)
    if hp.Boolean("dropout"):
        model.add(layers.Dropout(rate=0.25))
    # regression output
    out_reg = Dense(1, activation='linear')(hidden3)
    # classification output
    out_clas = Dense(n_class, activation='softmax')(hidden3)
    learning_rate = hp.Float("lr", min_value=1e-4, max_value=1e-2, sampling="log")

    model = Model(inputs=visible, outputs=[out_reg, out_clas])
    model.compile(loss=['mae','sparse_categorical_crossentropy'], optimizer=keras.optimizers.Adam(learning_rate=learning_rate),metrics=['mean_absolute_error','acc'])
    #plot_model(model, to_file='E:\model.png', show_shapes=True)
    #history=model.fit(X_train, [y_train,y_train_class],validation_split=0.4, epochs=50, batch_size=1000, verbose=2)
    #yhat1, yhat2 = model.predict(X_test)
    #error = mean_absolute_error(y_test, yhat1)
    #print('MAE: %.3f' % error)
    #yhat2 = argmax(yhat2, axis=-1).astype('int')
    #acc = accuracy_score(y_test_class, yhat2)
    #print('Accuracy: %.3f' % acc)
    return model
       
build_model(keras_tuner.HyperParameters())
```

```python
import math

def nrmse(y_true, y_pred):
 
    mse = ((y_true - y_pred) ** 2).mean()
    rmse = math.sqrt(mse)
    numerator = rmse
    denominator = max(y_true) - min(y_true)
    return numerator / denominator
```

```python
nrmse_score = nrmse(y_test,ypred[0])
print("NRMSE:", nrmse_score)
```
